```python
import matplotlib.pyplot as plt
import numpy as np

# Data for the funnel chart
stages = ['Stage 1', 'Stage 2', 'Stage 3', 'Stage 4', 'Stage 5']
values = [100, 80, 60, 40, 20]

# Calculate the cumulative percentages
total = sum(values)
percentages = [val / total for val in values]
cumulative_percentages = [sum(percentages[:i + 1]) for i in range(len(percentages))]

# Create the reversed funnel chart with padding on the left side
fig, ax = plt.subplots(figsize=(8, 6))

# Plot the reversed funnel segments
colors = plt.cm.get_cmap('Blues')(np.linspace(0.2, 1, len(stages)))
for i in range(len(stages)):
    ax.barh(i, percentages[i], color=colors[i], edgecolor='gray')

# Customize the funnel chart
ax.set_xlim([0, 1.1])  # Set the x-axis limits with padding on the left side
ax.set_ylim([len(stages) - 0.5, -0.5])  # Set the y-axis limits in reverse order
ax.set_yticks(range(len(stages)))  # Set the y-axis ticks
ax.set_yticklabels(stages[::-1], fontsize=12)  # Set the reversed y-axis labels
ax.set_xticks([])  # Remove x-axis ticks
ax.set_title('Reversed Funnel Chart with Left-side Padding', fontsize=14, fontweight='bold')

# Add data labels
for i in range(len(stages)):
    ax.text(percentages[i] + 0.02, len(stages) - 1 - i, f'{percentages[i] * 100:.1f}%', va='center', fontsize=12,
            fontweight='bold', color='black')

# Add cumulative percentages
for i in range(len(stages)):
    ax.text(0.05, len(stages) - 1 - i, f'{cumulative_percentages[i] * 100:.1f}%', ha='left', va='center', fontsize=12,
            fontweight='bold', color='black')

# Remove spines
ax.spines['top'].set_visible(False)
ax.spines['right'].set_visible(False)
ax.spines['left'].set_visible(False)

# Adjust layout
plt.tight_layout()

# Show the chart
plt.show()
```

```python
import psycopg2
import matplotlib.pyplot as plt
import numpy as np

# Establish a connection to the database
conn = psycopg2.connect(
    database="NewDB",
    user="postgres",
    password="Adduvaith789",
    host="127.0.0.1",
    port="5432"
)
cur = conn.cursor()

# Execute the SQL query
cur.execute("SELECT * FROM iris where sepallength>7")
rows = cur.fetchall()

labels = []
sizes = []

for row in rows:
    labels.append(row[0])
    sizes.append(row[1])

# Reverse the lists to start with the smallest value at the top
labels = labels[::-1]
sizes = sizes[::-1]

# Calculate the cumulative sum of sizes
cumulative_sizes = np.cumsum(sizes)

# Calculate the percentage for each step
percentages = 100 * cumulative_sizes / cumulative_sizes[-1]

# Plot the funnel chart
fig, ax = plt.subplots(figsize=(6, 8))
ax.barh(range(len(labels)), sizes, color='steelblue')
ax.set_yticks(range(len(labels)))
ax.set_yticklabels(labels)
ax.invert_yaxis()
ax.set_xlabel('Count')
ax.set_ylabel('Steps')

# Display the percentage labels
for i, p in enumerate(ax.patches):
    percentage = f'{percentages[i]:.1f}%'
    x = p.get_width() + 10
    y = p.get_y() + p.get_height() / 2
    ax.annotate(percentage, (x, y))

plt.show()

# Close the cursor and connection
cur.close()
conn.close()
```
